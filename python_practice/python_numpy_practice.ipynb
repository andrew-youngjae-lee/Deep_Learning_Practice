```python
from matplotlib import pyplot as plt

plt.plot([1,2,3],[1,4,9]) #plot에 여러개의 line을 추가 : plt.plot()을 여러번 사용하면 됨
plt.plot([1,2,3],[5,6,7])
plt.xlabel('Sequence')
plt.ylabel('Time(secs)')
plt.title('Experiment Result')
plt.legend(['Mouse', 'Cat']) #각 line에 대한 범례를 추가 : plt.legend(['라인1 범례', '라인2 범례']) - (순서대로 지정)
plt.show()
```
